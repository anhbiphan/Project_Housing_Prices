```python
# Helper function to collect avg. price for each zipcode
def zip_mean(zipcode):
    zip_mean = []
    for _zip in zipcode:
        _zip = df13m[(df13m['zipcode'] == _zip)]
        zip_mean.append(_zip['price'].mean().astype(int))
    return (zip_mean)
```

# Selecting Target and Features 

```python
# y is our prediction target
# choosing features to use 
y= df.price
df_features = ['bedrooms', 'bathrooms', 'sqft_lot', 'floors', 'sqft_living', 'sqft_living15',
               'sqft_lot15', 'lat', 'long', 'condition', 'sqft_above','grade','zipcode','lat', 'long']

X = df[df_features]
```

# Metric Mean Absolute Error and R2

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# Using Mean Absolute Error (can provide in dollars how far above or below the predictions are.)
# Using R2 score (can give us a % score of error)
```

# DTR Model Initial MAE/R2 scores

```python
# Training and Testing to predict then finding MAE
kc_model = DecisionTreeRegressor(random_state=1)
kc_model.fit(train_X, train_y)
val_predictions = kc_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
print(r2_score(val_y, val_predictions))
```

# Helper Function with Leaf_Nodes

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    val_preds = model.predict(val_X)
    mae = mean_absolute_error(val_y, val_preds)
    return (mae)
```

# Train / Test with diff. leaf nodes

```python
# Testing diff. leaf_nodes size for best results
# Grabbing lowest MAE score from helper function output
candid_leaf_nodes = [5, 25, 50, 100, 250, 500, 550, 600]
score = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candid_leaf_nodes}
best_leaf_size = min(score, key=score.get)
best_leaf_size
```

```python
# Using 550 leaf_nodes to give a better result than initial test
model = DecisionTreeRegressor(max_leaf_nodes=best_leaf_size, random_state=0)
model.fit(train_X, train_y)
val_preds = model.predict(val_X)
print(mean_absolute_error(val_y, val_preds))
print(r2_score(val_y, val_preds))
```

# Log Transform / Min-Max Scale

```python
import numpy as np
```

```python
# Log Transforming to reduce skewness of our data

data = pd.DataFrame([])
data['log_bathrooms'] = np.log(X['bathrooms'])
data['log_bedrooms'] = np.log(X['bedrooms'])
data['log_sqft_above'] = np.log(X['sqft_above'])
data['log_sqft_living'] = np.log(X['sqft_living'])
data['log_sqft_living15'] = np.log(X['sqft_living15'])
data['log_sqft_lot'] = np.log(X['sqft_lot'])
data['log_sqft_lot15'] = np.log(X['sqft_lot15'])

cond = X['condition']
floors = X['floors']
lat = X['lat']
long = X['long']
zipco = X['zipcode']
grade = X['grade']

# Scaling our data to normalize the values (Min-Max)

data['scaled_condition'] = (cond - min(cond)) / (max(cond) - min(cond))
data['scaled_floors'] = (floors - min(floors)) / (max(floors) - min(floors))
data['scaled_lat'] = (lat - min(lat)) / (max(lat) - min(lat))
data['scaled_long'] = (long - min(long)) / (max(long) - min(long))
data['scaled_zipcode'] = (zipco - min(zipco)) / (max(zipco) - min(zipco))
data['scaled_grade'] = (grade - min(grade)) / (max(grade) - min(grade))
data.hist(figsize=[13,13]);
```
