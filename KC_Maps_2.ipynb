```python
# import librarties
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import folium
from folium.plugins import HeatMap
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# make dataframe again
df=pd.read_csv('kc_house_data.csv')
```

```python
df['zipcode']=df['zipcode'].astype(str)
df['zipcode']
```

```python
df_low_price=df.drop(df[df.price>1300000].index)
df_low_price
```

```python
df_mid=df_low_price.drop(df_low_price[df_low_price.price<150000].index)
df_mid
```

```python
# building dataframe of averages for zipcodes
zipcode_data = df_mid.groupby('zipcode').aggregate(np.mean)
zipcode_data.reset_index(inplace=True)

# counting number of houses per zipcode
df_mid['count'] = 1
count_houses_zipcode = df_mid.groupby('zipcode').sum()
count_houses_zipcode.reset_index(inplace=True)
count_houses_zipcode = count_houses_zipcode[['zipcode','count']]
df_mid.drop(['count'], axis = 1, inplace = True)

# merge dataframes to make zipcode average dataframe
zipcode_data = pd.merge(zipcode_data, count_houses_zipcode, how='left', on=['zipcode'])
zipcode_data.head()
```

```python
# houses sold in zipcodes
def show_zipcode_map(col):
    # reading geojson file 
    geo_path = 'Zipcodes_for_King_County_and_Surrounding_Area_Shorelines__zipcode_shore_area.geojson'
    # basemap
    zipcode = folium.Map(location=[df_mid['lat'].mean(), df_mid['long'].mean()], zoom_start=9.4)
    # reading data to display zipcode polygons on basemap and fill with colors
    zipcode.choropleth(geo_data="Zipcodes_for_King_County_and_Surrounding_Area_Shorelines__zipcode_shore_area.geojson",
             data=zipcode_data,
             columns=['zipcode',col],
             key_on='feature.properties.ZIPCODE', 
             fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.2,legend_name='Number of Houses Sold per Zipcode')
    return zipcode
```

# Plotting individual homes on basemap with price ranges

```python
import matplotlib.cm
```

Function for making color range for houses

```python
  def ColorVector(values, number_of_colors, cmap_name):
      buckets = pd.qcut(values, number_of_colors).codes
      cmap = matplotlib.cm.get_cmap(name = cmap_name)
      colors = []
      for i in range(0, number_of_colors):
          r = int(cmap(i/number_of_colors)[0]*255) 
          g = int(cmap(i/number_of_colors)[1]*255)
          b = int(cmap(i/number_of_colors)[2]*255)
          color = "#{0:02x}{1:02x}{2:02x}".format(r, g, b)  
          colors.append(color)
      values = []
      for j in range(0, len(buckets)):
          value = colors[buckets[j]]
          values.append(value)
      return values
```

```python
# function for splitting range of colors
def ColorCategories(variable, colors):
    df['Colors'] = colors
    counts = df.groupby(['Colors', variable]).size().reset_index().rename(columns={0:'count'})
    color_list = counts.Colors.unique()
    for color in color_list:
        temp = counts[(counts.Colors == color)]
        print(color, min(temp[variable]), max(temp[variable]), len(temp))
```

```python
lats = df['lat']
lngs = df['long']
price=df['price']
sqft = df['sqft_living']
```

```python
dollar_vals = []
for i in range(0, len(price)):
    dollar_val = '${:0,.0f}'.format(price[i]).replace('$-','-$')
    
# stylize as number (commas)
num_sqfts = []
for i in range(0, len(sqft)):
    num_sqft = '{:0,.0f}'.format(sqft[i])
    num_sqfts.append(num_sqft)
    
# determine color vector    
colors = ColorVector(df['price'].values, 10, 'plasma')
ColorCategories('price', colors)

# create the base map
m = folium.Map(location=[np.mean(df.lat), np.mean(df.long)], zoom_start = 9.4)

# loop through properties to add map layers
for i in range(len(lats)):
    folium.Circle(location=[lats[i], lngs[i]], 
                  #popup = ("""<b>{}</b><br>{}<br>{} sq. ft.<br>""".format(addresses[i], dollar_vals[i], num_sqfts[i])),
                  radius = 20, 
                  color = colors[i],
                  fill = True,
                  fill_color = colors[i], 
                  fill_opacity=.30).add_to(m)
```

```python
# overlay zipcode polygons on basemap
m.choropleth(geo_data="Zipcodes_for_King_County_and_Surrounding_Area_Shorelines__zipcode_shore_area.geojson",
             data=df_low_price, # my dataset
             columns=['zipcode', 'price'], # zip code is here for matching the geojson zipcode, sales price is the column that changes the color of zipcode areas
             key_on='feature.properties.ZIPCODE', # this path contains zipcodes in str type, this zipcodes should match with our zipcode column
             fill_color='BuGn', fill_opacity=0, line_opacity=0.5)
```
